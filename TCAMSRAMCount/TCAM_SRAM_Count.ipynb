```python
import re
```

```python
def check(string,c) :
    regex = re.compile("exact")
    match_object = regex.findall(string)
    for word in match_object:
        print(word)
        c=c+1
    print(c)
```

```python
if __name__ == '__main__' :
  
    # Enter the string
    for line in Lines:
        string =line
  
    # Calling check function
        check(string,count)
```
