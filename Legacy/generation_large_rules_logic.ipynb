```python
import pandas as pd
import json
df = pd.read_csv (r'E:/PhD/data_output/each_device_flow-rule/amazon_echo.csv') 
c= "9e:8d:de:80:29:28"
length=len(c)
print(length)
for element in range(0, len(c)):
    if element==length-1:
        temp=c[element]
        #print(temp)
i=0
while i<74:
   
    #var1="9e:8d:de:80:29:"+temp
    #print(var1)
    #print(temp1)
    if i<=9:
        temp1=temp+str(i)
        df2=df.replace(to_replace ="9e:8d:de:80:29:28",value ="9e:8d:de:80:29:"+temp1)   
    else:
        df2=df.replace(to_replace ="9e:8d:de:80:29:28",value ="9e:8d:de:80:"+temp1+":"+str(i)) 
    if i==0:
        result=df.append(df2)
    if i>0:
        result=df2.append(result)
    i=i+1
result = result[['sEth','dEth','typEth','proto','sPort','dPort','Source','Destination','action']]
result.to_csv('E:/PhD/data_output/70k/amazon_echo.csv',index=None,header=True)
```
